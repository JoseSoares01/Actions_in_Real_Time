```python
import pandas as pd

# Assuming 'dataset' is your DataFrame

# Check data types of columns
print(dataset[['Age', 'Weight']].dtypes)

# Convert 'Age' and 'Weight' to numeric if they are not already
dataset['Age'] = pd.to_numeric(dataset['Age'], errors='coerce')
dataset['Weight'] = pd.to_numeric(dataset['Weight'], errors='coerce')

# Drop rows with missing values in 'Age' or 'Weight' if necessary
dataset.dropna(subset=['Age', 'Weight'], inplace=True)

# Now create the histograms
dataset.hist(['Age', 'Weight'], bins=50)
```
